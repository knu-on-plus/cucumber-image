```python
import os
import json

# YOLO ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
model = YOLO("/home/oi/Desktop/song/cucumber-image/yolo_detection/oi/weights/best.pt")
# train ë° val ì´ë¯¸ì§€ ê²½ë¡œ
train_dir = '/home/oi/Desktop/song/cucumber-image/data/train'
val_dir = '/home/oi/Desktop/song/cucumber-image/data/val'
label_dir = '/home/oi/Desktop/song/cucumber-image/data/labels'  # ê¸°ì¡´ ë ˆì´ë¸” JSON íŒŒì¼ì´ ìžˆëŠ” ë””ë ‰í† ë¦¬

# YOLO ì¶”ë¡  ë° ë ˆì´ë¸” ì—…ë°ì´íŠ¸ í•¨ìˆ˜
def update_labels_with_yolo(model, image_dir, label_dir):
    total_new_labels = 0  # ìƒˆë¡œ ì¶”ê°€ëœ ë ˆì´ë¸” ìˆ˜ë¥¼ ì €ìž¥í•  ë³€ìˆ˜
    
    for image_file in os.listdir(image_dir):
        if image_file.endswith(('.jpg', '.jpeg', '.png')):  # ì´ë¯¸ì§€ íŒŒì¼ë§Œ ì²˜ë¦¬
            image_path = os.path.join(image_dir, image_file)
            label_file = os.path.splitext(image_file)[0] + '.json'
            label_path = os.path.join(label_dir, label_file)
            
            # ê¸°ì¡´ ë ˆì´ë¸” íŒŒì¼ ì—´ê¸°
            if os.path.exists(label_path):
                with open(label_path, 'r', encoding='utf-8') as f:
                    label_data = json.load(f)
            else:
                continue  # ë ˆì´ë¸” íŒŒì¼ì´ ì—†ìœ¼ë©´ ë‹¤ìŒ ì´ë¯¸ì§€ë¡œ ì´ë™
            
            # ê¸°ì¡´ ë ˆì´ë¸”ì˜ ê°ì²´ ìˆ˜
            existing_objects = len(label_data['result'])
            
            # YOLO ì¶”ë¡  ìˆ˜í–‰
            results = model(image_path)
            detections = results[0].boxes
            # ì´ë¯¸ì§€ í¬ê¸° ì •ë³´ ê°€ì ¸ì˜¤ê¸°
            image_width = label_data['description']['width']
            image_height = label_data['description']['height']
            
            # YOLO íƒì§€ ê²°ê³¼ë¥¼ ê¸°ì¡´ ë ˆì´ë¸”ì— ì¶”ê°€
            for detection in detections:
                xyxy = detection.xyxy[0].cpu().numpy()  # ê²½ê³„ ìƒìž ì¢Œí‘œ [x_min, y_min, x_max, y_max]
                conf = detection.conf.cpu().numpy()  # ì‹ ë¢°ë„
                cls = detection.cls.cpu().numpy()  # í´ëž˜ìŠ¤
                
                # ê²½ê³„ ìƒìž ì¢Œí‘œë¥¼ ë³€í™˜í•˜ì—¬ JSON í˜•ì‹ì— ë§žê²Œ ì €ìž¥
                x_min, y_min, x_max, y_max = xyxy
                x = int(x_min)
                y = int(y_min)
                w = int(x_max - x_min)
                h = int(y_max - y_min)
                
                # ìƒˆë¡œ íƒì§€ëœ ê°ì²´ë¥¼ ë ˆì´ë¸”ì— ì¶”ê°€
                new_bbox = {
                    "id": "generated_id",  # ê³ ìœ í•œ ID
                    "type": "bbox",
                    "x": x,
                    "y": y,
                    "w": w,
                    "h": h,
                    "cx": None,
                    "cy": None,
                    "r": None,
                    "segmentation": []
                }
                label_data['result'].append(new_bbox)
                print(f'x: {x}, y: {y}, w: {w}, h: {h}')
            # ìƒˆë¡œ ì¶”ê°€ëœ ê°ì²´ ìˆ˜ ê³„ì‚°
            new_objects = len(label_data['result']) - existing_objects
            total_new_labels += new_objects
            
            # # ì—…ë°ì´íŠ¸ëœ ë ˆì´ë¸” íŒŒì¼ ì €ìž¥
            # with open(label_path, 'w') as f:
            #     json.dump(label_data, f, indent=4)
            break
    return total_new_labels

# train ë° val ì´ë¯¸ì§€ì— ëŒ€í•´ YOLO ì¶”ë¡  ë° ë ˆì´ë¸” ì—…ë°ì´íŠ¸ ìˆ˜í–‰
new_train_labels = update_labels_with_yolo(model, train_dir, label_dir)
new_val_labels = update_labels_with_yolo(model, val_dir, label_dir)

# ìƒˆë¡œ ì¶”ê°€ëœ ë ˆì´ë¸” ìˆ˜ ì¶œë ¥
print(f"Train setì—ì„œ ìƒˆë¡­ê²Œ ì¶”ê°€ëœ ë ˆì´ë¸” ìˆ˜: {new_train_labels}")
print(f"Validation setì—ì„œ ìƒˆë¡­ê²Œ ì¶”ê°€ëœ ë ˆì´ë¸” ìˆ˜: {new_val_labels}")
print(f"ì´ ìƒˆë¡­ê²Œ ì¶”ê°€ëœ ë ˆì´ë¸” ìˆ˜: {new_train_labels + new_val_labels}")
```
